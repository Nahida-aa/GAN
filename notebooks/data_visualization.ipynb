```python
import os
import pandas as pd
import matplotlib.pyplot as plt

#  CSV 
def read_and_merge_csv_files(log_dir, file_names):
    df_list = []
    for file_name in file_names:
        file_path = os.path.join(log_dir, file_name)
        df = pd.read_csv(file_path, header=0, names=['Timestamp', 'Epoch', 'Batch', 'D Loss', 'G Loss'], skiprows=1)
        df_list.append(df)
    merged_df = pd.concat(df_list, ignore_index=True)
    return merged_df

#  NaN 
def preprocess_dataframe(df):
    df['Epoch'] = pd.to_numeric(df['Epoch'], errors='coerce')
    df['Batch'] = pd.to_numeric(df['Batch'], errors='coerce')
    df['D Loss'] = pd.to_numeric(df['D Loss'], errors='coerce')
    df['G Loss'] = pd.to_numeric(df['G Loss'], errors='coerce')
    df = df.dropna()
    return df

# 
def exponential_moving_average(data, span):
    return data.ewm(span=span, adjust=False).mean()

# 
def smooth_data(df, span):
    df['D Loss Smoothed'] = exponential_moving_average(df['D Loss'], span)
    df['G Loss Smoothed'] = exponential_moving_average(df['G Loss'], span)
    return df


# 
def plot_loss(df, title, ylabel, column, smoothed_column):
    plt.figure(figsize=(12, 6))
    plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df[smoothed_column], label=column)
    plt.xlabel('Epoch')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
    
def main():
    #  CSV 
    BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
    log_dir = os.path.join(BASE_DIR, 'output/logs')
    file_names = [
        'training_20241124-192707.log.csv',
        'training_20241124-195411.log.csv',
        'training_20241124-195843.log.csv',
        'training_20241124-201740.log.csv',
        'training_20241124-205454.log.csv',
    ]  # 
    df = read_and_merge_csv_files(log_dir, file_names)

    # 
    print(df.head())

    # 
    df = preprocess_dataframe(df)

    # 
    span = 468 * 50  # EMA 
    df = smooth_data(df, span)

    # 
    plot_loss(df, 'Discriminator Loss Over Time', 'D Loss', 'D Loss', 'D Loss Smoothed')

    # 
    plot_loss(df, 'Generator Loss Over Time', 'G Loss', 'G Loss', 'G Loss Smoothed')

    # 
    plt.figure(figsize=(12, 6))
    plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df['D Loss Smoothed'], label='D Loss')
    plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df['G Loss Smoothed'], label='G Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Discriminator and Generator Loss Over Time')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
