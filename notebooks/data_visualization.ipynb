```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..'))
log_dir = os.path.join(BASE_DIR, 'output/logs')
csv_file_path = os.path.join(log_dir, 'training_20241123-010528.log.csv')
df = pd.read_csv(csv_file_path)

# 查看数据框架的前几行
print(df.head())

# 定义移动平均函数
def moving_average(data, window_size):
    return data.rolling(window=window_size).mean()

# 平滑数据
window_size = 50
df['D Loss Smoothed'] = moving_average(df['D Loss'], window_size)
df['G Loss Smoothed'] = moving_average(df['G Loss'], window_size)

# 绘制判别器损失曲线
plt.figure(figsize=(12, 6))
plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df['D Loss Smoothed'], label='D Loss')
plt.xlabel('Epoch')
plt.ylabel('D Loss')
plt.title('Discriminator Loss Over Time')
plt.legend()
plt.show()

# 绘制生成器损失曲线
plt.figure(figsize=(12, 6))
plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df['G Loss Smoothed'], label='G Loss')
plt.xlabel('Epoch')
plt.ylabel('G Loss')
plt.title('Generator Loss Over Time')
plt.legend()
plt.show()

# 绘制判别器和生成器损失曲线在同一图中
plt.figure(figsize=(12, 6))
plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df['D Loss Smoothed'], label='D Loss')
plt.plot(df['Epoch'] + df['Batch'] / df['Batch'].max(), df['G Loss Smoothed'], label='G Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Discriminator and Generator Loss Over Time')
plt.legend()
plt.show()
```
